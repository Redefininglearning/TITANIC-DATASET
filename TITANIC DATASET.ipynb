CLEANING THE DATASET

```python
# Checking the sum of duplicates
TITANIC.duplicated().sum
```

```python
# Checking for duplicates in the dataset
TITANIC.drop_duplicates()
```

```python
# Dropping of columns that will not be useful
TITANIC.drop(['PassengerId','Ticket','Cabin','Embarked','Name'],axis=1,inplace =True)
```

```python
TITANIC
```

```python
# Filling the missing values in 'Age' column using median
median_Age= TITANIC['Age'].median()
median_Age
```

```python
TITANIC['Age'].fillna(median_Age,inplace = True)
```

```python
TITANIC
```

```python
# Converting float to integer under 'Age' column
TITANIC['Age']= TITANIC['Age'].astype(int)
```

```python
TITANIC
```

```python
# Confirming if there are still empty cells in the dataset
TITANIC.isnull().sum()
```

```python
# Converting categorical variables in 'Sex' column to continuous variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
TITANIC['Sex'] =le.fit_transform(TITANIC['Sex'])
```

```python
TITANIC
```

 DIVIDING THE DATASET INTO INDEPENDENT AND DEPENDENT VARIABLES

```python
from sklearn.model_selection import train_test_split
X=TITANIC.drop(columns='Survived')
Y= TITANIC['Survived']
```

```python
X
```

```python
Y= TITANIC['Survived']
Y
```

```python
X_train, X_test,y_train,y_test = train_test_split(X, Y, test_size=0.3, random_state=45)
```

```python
X_train
```

```python
# Train Decision Tree model
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
X_test
```

```python
Predictions= model.predict(X_test)
Predictions
```

```python
#  Using Evaluation Metrics to find accuracy score
from sklearn.metrics import accuracy_score
score =accuracy_score(y_test,Predictions)
```
